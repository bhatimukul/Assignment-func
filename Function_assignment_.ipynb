Q6. Explain the concept of generators in python and how they are defined.

Ans. In Python, generators are a type of iterable that allow you to iterate over data without storing the entire sequence in memory at once. They are useful when dealing with large datasets or streams of data because they yield items one at a time, on demand, rather than producing the entire collection all at once.

Key Features of Generators:
1.	Lazy Evaluation: Generators produce values only when needed, conserving memory and improving performance for large data.
2.	State Retention: Generators automatically retain their state between executions, so each call to the generator resumes where it last left off.
3.	Yield vs. Return: A generator uses the yield keyword instead of return. While return exits a function and provides a value, yield pauses the function and provides a value while maintaining the function's state so it can be resumed later.

How to Define a Generator:

Generators are defined just like regular functions but use the yield keyword instead of return.

Using a Generator:

You can iterate over a generator using a loop or the next() function.

Generator Expressions:

You can also create generators using generator expressions, which are similar to list comprehensions but use parentheses instead of square brackets.

Advantages of Generators:

•	Memory Efficiency: Only one item is stored at a time.

•	Performance: Ideal for large datasets or infinite sequences where you don't need all elements at once.

•	Composability: Generators can be combined, transformed, or filtered without materializing the entire sequence.

Generators are powerful for scenarios where lazy, on-demand computation is preferable

5.	Simpler Code: Generators can often lead to more concise and readable code compared to using iterators or managing state manually.
6.	Infinite Sequences: Generators can represent infinite sequences, such as a stream of data or numbers, because they generate values on-the-fly rather than requiring a fixed-size structure.

Q10.What is the difference between ‘map()’, ‘reduce()’, and filter()’ functions in python?
Ans. In Python, map(), filter(), and reduce() are built-in functions that help apply a function to iterables like lists or tuples. They are part of functional programming and allow you to manipulate collections in a concise way. Here's the difference between them:
1. map():

•	Purpose: Applies a function to every item in an iterable (e.g., list, tuple) and returns a new iterable (map object) with the results.

•	Usage: When you want to transform each element of an iterable by applying a function.

•	Syntax: map(function, iterable)

2. filter():

•	Purpose: Filters elements from an iterable by applying a function that returns True or False. It returns a new iterable (filter object) with only the elements for which the function returned True.

•	Usage: When you want to extract elements from an iterable that satisfy a condition.

•	Syntax: filter(function, iterable)

3. reduce():

•	Purpose: Applies a binary function (a function that takes two arguments) cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value.

•	Usage: When you want to combine or aggregate all the elements of an iterable into a single value.

•	Syntax: reduce(function, iterable)

•	Note: reduce() is not a built-in function in Python 3, so you need to import it from the functools module.

Summary:

•	map(): Transforms each item in an iterable.

•	filter(): Selects items from an iterable that meet a condition.

•	reduce(): Combines all items in an iterable into a single result.
Each function has a specific use case, making it easy to manipulate collections functionally.

```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
numbers = [1, 2, 3, 4, 5, 6]
print(sum_of_even_numbers(numbers))  # Output: 12

# This function uses a list comprehension to filter even numbers and sums them up
```

Q4. write a python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:  # Prime numbers are greater than 1
        return False
    if n == 2:  # 2 is the only even prime number
        return True
    if n % 2 == 0:  # Exclude other even numbers
        return False
    for i in range(3, int(n ** 0.5) + 1, 2):  # Check divisibility from 3 up to √n, skipping evens
        if n % i == 0:
            return False
    return True

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(f"{number} is a prime number")
```

Q5. Create  an iterator class in python that generates the fibonacci sequence up to a specified number of terms.



```python
class FibonacciIterator:
    def __init__(self, num_terms):
        self.num_terms = num_terms  # Number of terms to generate
        self.current_term = 0       # Tracks the current term index
        self.a, self.b = 0, 1       # Starting values of the Fibonacci sequence

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.num_terms:  # Stop iteration after reaching the specified number of terms
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.a  # Return the first Fibonacci number (0)
        elif self.current_term == 1:
            self.current_term += 1
            return self.b  # Return the second Fibonacci number (1)
        else:
            self.a, self.b = self.b, self.a + self.b  # Calculate the next Fibonacci number
            self.current_term += 1
            return self.b

# Example usage:
fib_iter = FibonacciIterator(10)  # Create an iterator for the first 10 terms
for num in fib_iter:
    print(num)

# Explanation:
# __iter__(self): This method makes the class an iterator, returning itself.
# __next__(self): This method defines the logic for generating the next Fibonacci number. It raises StopIteration once the specified number of terms is reached.
```

Q6. write a generator function in python that yields the power of 2 up to a given exponent.


```python
def power_of_two(max_exponent):
    for exp in range(max_exponent + 1):
        yield 2 ** exp

# Example usage:
for power in power_of_two(5):
    print(power)

```

Q9.write a python program that uses 'map()'to convert a list of temperatures from celsius to fahrenheit.


```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Use map() to convert the list to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Example output
print(fahrenheit_temperatures)
[32.0, 68.0, 98.6, 212.0]
```
